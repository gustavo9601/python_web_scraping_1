```python
soup_latam = BeautifulSoup(request_url.text, 'lxml')
soup_latam.prettify()
```

```python
# Argumentos para navegador
options = webdriver.ChromeOptions()
# Acceder a modo incognito
options.add_argument('--incognito')

# Intanciando driver para selenium
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)
# Se abre el nabegador en la url pasada por parametro
driver.get(url)

# Cerrando el navegador
# driver.close()
```

```python
# Usano a traves de xpath
vuelos = driver.find_elements_by_xpath('//li[@class="sc-gQNndl cutlUv"]')
vuelos
```

```python
# .// Indica que busque a partir del nodo seleccionado anteriormente
hora_salida_vuelo_1 = vuelos[0].find_elements_by_xpath('.//div[@class="sc-bOCYYb bpIvuu"]//div[@class="sc-iFUGim fupCEn flight-information"]/span[@class="sc-clBsIJ kaUpli"]')[0].text
hora_regreso_vuelo_1 = vuelos[0].find_elements_by_xpath('.//div[@class="sc-bOCYYb bpIvuu"]//div[@class="sc-iFUGim fupCEn flight-information"]/span[@class="sc-clBsIJ kaUpli"]')[1].text


print(hora_salida_vuelo_1, hora_regreso_vuelo_1)
```
