```python
# Imports
import requests
from IPython.display import Image
from bs4 import BeautifulSoup
```

```python
URL = 'https://www.pagina12.com.ar/'
```

```python
# Descarga la pagina de la URL
page12 = requests.get(URL)

# Se verifica el stado, y ya obtememos el html
print(page12.status_code)

# Se puede ver el texto en raw del html
print(page12.text)
```

```python
# Muestra el encabezado con el que sale la respuesta
print(page12.request.headers)

# Muestra el encabezado de la respuesta
print(page12.headers)

# Muestra a que url le realiza la solicitud
print(page12.request.url)
```

```python
# Parseando el codeigo HTML

soup = BeautifulSoup(page12.text, 'lxml')
# Denta el codigo Html
soup.prettify()

```

```python
# Extrayendo datos del html parseado
# .find('tag') // busca el primer elemento
# .find_all('tag') // Retorna una lista con los elementos encontrados
secciones = soup.find('ul', attrs={'class': 'horizontal-list main-sections hide-on-dropdown'}).find_all('li')

# .get('atributo') // optiene algun atributo
# .get_text() // optiene el texto del elemento
secciones[0].find('a').get('href')
secciones[0].find('a').get_text()
```

```python
# Alamacenando los textos y links obteniedos
links_secciones = [seccion.find('a').get('href') for seccion in secciones]
texto_secciones = [seccion.find('a').get_text() for seccion in secciones]
links_secciones
```

```python
# Accediendo al primer link encontrado y extrayendo los articulos
"""
Lo habitual es capturar la expecion al usar peticiones a terceros
try:
    pass
except Exeption as e:
    pass


"""
# Descarga la pagina de la URL
page12_articles = requests.get(links_secciones[0])
print(page12_articles.status_code)

# Parsenado el html
soup_seccion_article = BeautifulSoup(page12_articles.text, 'lxml')

# Obteniendo el articulo principal
featured_article = soup_seccion_article.find('article', attrs={'class': 'article-item article-item--main'})
link_article = featured_article.find('h2', attrs={'class': 'title-list'}).find('a').get('href')
link_article
```

```python
# Extrayendo los demas links de las demas listas
articles_list = soup_seccion_article.find('div', attrs={'class': 'articles-list'}).find_all('div', attrs={
    'class': 'article-item__header'})

url_notas = []
for article in articles_list:
    url_notas.append(article.find('a').get('href'))

print(url_notas)
```

```python
# Obteniendo la informacion para las notas

try:
    nota = requests.get(url_notas[0])
    if nota.status_code == 200:
        soup_nota = BeautifulSoup(nota.text, 'lxml')

        title = soup_nota.find('h1').getText()
        fecha = soup_nota.find('span', attrs={'pubdate': 'pubdate'}).get('datetime')
        image = soup_nota.find('figure', attrs={'class': 'object-fit-block--contain intrinsic-container intrinsic-container-3x2'}).find_all('img')[0].get('data-src')

        # Download image
        img_req = requests.get(image)
        # Permite rendedirazr el contenido en Jupiter
        # content accede al contenido binario del archivo
        Image(img_req.content)


except Exception as error:
    print('Error: ')
    print(error)
    print('\n\n')
```
